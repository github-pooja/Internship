# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.


```python
#Requesting the webpage server to get the source code
```

```python
imdb=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc", verify=False)
imdb
```

```python
#Now we'll parse the source code by using Beautiful Soup.
```

```python
soup=BeautifulSoup(imdb.content)
soup
```

```python
# Now we need to extract the  Name, IMDB Ratings,Year of Release of movies.
```

```python
imdb_movies=soup.find_all('h3',class_='lister-item-header')
imdb_movies
```

```python
top_rated=[]
```

```python
for i in imdb_movies:
    top_rated.append(i.get_text())
top_rated
```

```python
top_rated=[]
for i in imdb_movies:
    top_rated.append(i.get_text().strip().split('\n')[1].strip())
top_rated
```

```python
#Here we are extracting teh Year of Release
```

```python
movies_yr=soup.find_all('span',class_="lister-item-year text-muted unbold")
movies_yr
```

```python
# Now Extracting the year from these tags using for loop
```

```python
year_list=[]
```

```python
for i in movies_yr:
    year_list.append(i.get_text())
year_list
```

```python
year_list=[]
for i in movies_yr:
    year_list.append(i.text.replace('(','').replace(')',''))
year_list
```

```python
#Now we will extract the Movies Rating
```

```python
ratings=soup.find_all('div',class_="inline-block ratings-imdb-rating")
ratings
```

```python
rating_list=[]
```

```python
for i in ratings:
    rating_list.append(i.text)
rating_list
```

```python
rating_list=[]
for i in ratings:
    rating_list.append(i.text.replace('\n',''))
rating_list
```

```python
#Hence we have extracted the required data as shown above now we'll be checking the length of all lists before performing the dataframe formation.
```

```python
print(len(top_rated),len(year_list),len(rating_list))
```

```python
#Hence its confirm that all the lists have equal length so now we can perform dataframe.
```

```python
import pandas as pd
imdb=pd.DataFrame({})
imdb['Movie_Name']=top_rated
imdb['Year_of_Release']=year_list
imdb['Rating']=rating_list
imdb.head()
```

```python
#Saving to csv
#Solution2
imdb.to_csv("Top_Rated_100_movies.csv")
```

# Ques.3 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file. 


```python
#Requesting the webpage server to get the source code
```

```python
indian_movies=requests.get("https://www.imdb.com/india/top-rated-indian-movies/",verify=False)
```

```python
indian_movies.content
```

```python
#Now we'll parse this souce code by using beautifulsoup. The parser used here is html.parser as the souce code present is in HTML.
```

```python
soup=BeautifulSoup(indian_movies.content)
soup
```

```python
# Now we'll extract all required data from this source code.
```

```python
#We'll extract the all the tags having the name of top rated indian movies.
```

```python
names=soup.find_all('td',class_='titleColumn')
names
```

```python
#Extracting text from above tags by using for loop
```

```python
movie_name=[]
for i in names:
    movie_name.append(i.text)
movie_name
```

```python
#Now we'll use split to separate the movie names and its year of release
```

```python
movie_name=[]
year_of_release=[]
for i in names:
    movie_name.append(i.get_text().strip().split('\n')[1].strip())
    year_of_release.append(i.get_text().strip().split('\n')[2].strip().replace('(','').replace(')',''))
movie_name
```

```python
year_of_release
```

```python
#Now we'll extract the rating of movies.
```

```python
ratings=soup.find_all('td',class_='ratingColumn imdbRating')
ratings
```

```python
#using for loop to get movies ratings
```

```python
movie_ratings=[]
for i in ratings:
    movie_ratings.append(i.text.replace('\n',''))
movie_ratings
```

```python
#So we got all reaquired list of indian movies now we'll perform the Dataframe.
```

```python
#checking the length of all lists.
print(len(movie_name),len(year_of_release),len(movie_ratings))
```

```python
#importing pandas
import pandas as pd
top_indian_movies=pd.DataFrame({})
top_indian_movies['movie_name']=movie_name
top_indian_movies['Releasing_year']=year_of_release
top_indian_movies['Ratings']=movie_ratings
```

```python
top_indian_movies
```

```python
#Displaying top 100 indaian movies
top_indian_movies.head(100).to_csv("Top_rated_100_Indian_Movies.csv")
```

# ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
odi_mens=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting",verify=False)
odi_mens
```

```python
odi_mens.content
```

```python
#Parsing the source code using BeautifulSoup
soup=BeautifulSoup(odi_mens.content)
```

```python
soup
```

```python
#Now we are scraping the top Batsmen's Name, Ratings, Teams 
```

```python
data = []
table = soup.find("table", class_="table rankings-table")
rows = table.findAll('tr')
for row in rows:
    cols = row.find_all('td')[0:4]
    cols = [ele.text.strip() for ele in cols]
    data.append([ele.replace('\n','').replace('This player has moved up in the rankings since the previous rankings update','').replace('This player has moved down in the rankings since the previous rankings update','').replace(' ','') for ele in cols if ele])
data[1:11]
```

# Ques.8) Write a python program to extract information about the local weather from the National
Weather Service website of USA, https://www.weather.gov/ for the city, San
Francisco. You need to extract data about 7 day extended forecast display for the city.
The data should include period, short description, temperature and description.

```python
weather=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJg99rUzY2w",verify=False)
weather
```

```python
soup=BeautifulSoup(weather.content)
soup
```

```python
period=soup.find_all('p',class_='period-name')
period
```

```python
#Now we'll extract required data from above tag using for loop
```

```python
period_list=[]
for i in period:
    period_list.append(i.text)
period_list
```

```python
short_desc=soup.find_all('p',class_="short-desc")
short_desc
```

```python
short_desc_list=[]
for i in short_desc:
    short_desc_list.append(i.text)
short_desc_list
    
```

```python
temp=soup.find_all('div',class_="tombstone-container")
temp
```

```python
temp_list=[]
for i in temp:
    temp_list.append(i.text.split(':')[1])
temp_list
    
```

```python
desc=soup.find_all('div',class_="col-sm-10 forecast-text")[0:9]
desc
```

```python
detailed_desc=[]
for i in desc:
    detailed_desc.append(i.text)
detailed_desc
    
```

```python
print(len(detailed_desc),len(short_desc_list),len(temp_list),len(period_list))
```

```python
# As we got all relevant information so now we can frame the dataframe for better look.
```

```python
import pandas as pd
```

```python
temp_forecast=pd.DataFrame({})
temp_forecast['Period']=period_list
temp_forecast['Temp']=temp_list
temp_forecast['Short_desc']=short_desc_list
temp_forecast['Detailed_desc']=detailed_desc
```
