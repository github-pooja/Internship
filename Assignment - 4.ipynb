```python
#Creating Dataframe to store the data
BCCI=pd.DataFrame({})
BCCI = pd.DataFrame({"Match":match_title,"Series":series,"Place":place,
                "Date":date,"Time":time})
print("\nTeam India’s International Fixtures: ")
BCCI
```

```python
#Creating empty list:
Repo_Title=[]
Repo_Desc=[]
URLs =[] #all urls
Languages=[]

#Getting Repository title element
Repo_title_elements=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
count=len(Repo_title_elements)
for i in range(count):
    #Extracting url repository:
    URLs.append(Repo_title_elements[i].get_attribute('href'))
    
for i in Repo_title_elements:#Appending the text in Repo_Title List from all elements
    Repo_Title.append(i.text)
Repo_Title


```

```python
Repo_Description=[]
for i in URLs:
    driver.get(i)
    try:
        repo_desc_element=driver.find_element_by_xpath("//p[@class='f4 mt-3']")#scraping description from the xpath
    #description.append(i.text)#appending the repository description in list
        Repo_Description.append(repo_desc_element.text)
    except NoSuchElementException:
        Repo_Description.append('--')
    time.sleep(2)
Repo_Description
```

```python
Contributor_count=[]
for i in URLs:
    driver.get(i)
    try:
        contri_count_element=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")#scraping contributors count from the xpath
        #appending the contributors count in list
        if contri_count_element.text is None:
             Contributor_count.append('--')
        else:
             Contributor_count.append(contri_count_element.text)
    except NoSuchElementException:
        Contributor_count.append('--')
    time.sleep(2)
Contributor_count
```

```python
Languages=[]

for i in URLs:
    driver.get(i)
    try:
        language_string=''   #for storing languages more than one
        language_element_span=driver.find_elements_by_xpath("//div[@class='BorderGrid-cell']/ul[@class='list-style-none']/li[@class='d-inline']/a/span")  #Scrapping used languages
        if(len(language_element_span)!=0):
            for j in language_element_span:
                 if '%' not in j.text:
                        language_string= language_string+','+ j.text
            Languages.append(language_string[1: len(language_string)])  #appending languages in list
        else:
              Languages.append("--")
    except NoSuchElementException:
        Languages.append('--')
    time.sleep(2)
Languages
```

```python
#Connecting to web driver
driver = webdriver.Chrome(r"C:\Users\Pooja.Mishra\Downloads\chromedriver_win32\chromedriver.exe")

#Hitting the url imdb.com.....
url = 'https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
#Finding elements name of shows:
imdb_series_elements=driver.find_elements_by_xpath("//div[@class='lister-item-content']")   #Findling common elementcontaining maximum required data elements.
#print(len(imdb_series_elements)) #imdb series complete data elements

#creating empty lists:
series_Name=[]
Year_Span=[]
Genre=[]
RunTime=[]
Ratings=[]
Votes=[]

for i in imdb_series_elements:
    year_element=i.find_elements_by_xpath(".//h3[@class='lister-item-header']/span")[1]  #Year_Span
    Year_Span.append(year_element.text)
    
    series_Name_element  = i.find_element_by_xpath(".//h3[@class='lister-item-header']/a")  #Series Name
    series_Name.append(series_Name_element.text)
    
    try:
        genre_element  = i.find_element_by_xpath(".//span[@class='genre']")   #Genre
        #print(genre_element.text)
        Genre.append(genre_element.text)
    except NoSuchElementException:
        Genre.append("--")
    
    try:
        runtime_element  = i.find_element_by_xpath(".//span[@class='runtime']")   #Runtime
        RunTime.append(runtime_element.text)
    except NoSuchElementException:
        RunTime.append("--")
        
    try:
        ratings_element  = i.find_element_by_xpath(".//span[@class='ipl-rating-star__rating']")   #Runtime
        Ratings.append(ratings_element.text)
    except NoSuchElementException:
        Ratings.append("--")
        
    try:
        votes_element  = i.find_element_by_xpath(".//span[@name='nv']")   #Votes
        Votes.append(votes_element.text)
    except NoSuchElementException:
        Votes.append("--")

        #Checking the length of data        
print (len(series_Name))
print (len(Year_Span))
print (len(Genre))
print (len(RunTime))
print (len(Ratings)) 
print (len(Votes))    
```

```python
#Creating dataframe to store the data:

import pandas as pd
IMDB = pd.DataFrame({})
IMDB["Name"] = series_Name
IMDB["Year Span"] = Year_Span
IMDB["Genre"] = Genre
IMDB["RunTime"] = RunTime
IMDB["Ratings"] = Ratings
IMDB["Votes"] = Votes
print("\nTop 100 most watched TV Series: ")
IMDB    #Printing the Dataframe
```
