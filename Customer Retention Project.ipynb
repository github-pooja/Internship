# Data Analysis

```python
# Combining all categorical data

cat_features=[features for features in cr.columns if cr[features].dtypes=="O"]
print(len(cat_features))
cr[cat_features].head()
```

```python
#Analysing the data on basis of city for online shopping
plt.figure(figsize=(20,5))
ax=sns.countplot(x=cr['3 Which city do you shop online from?'],linewidth=1)
print(cr['3 Which city do you shop online from?'].value_counts())
```

```python
for feature in cat_features:  
    data=cr.copy()
    plt.figure(figsize=(20,5))
    data[feature].hist(bins=30)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

Conclusions from above Graphs:
- 1) Which E-com Website offers wild variety of products?
From above graph we can conclude that Amazon.in+Flipkart.com have maximum offers than others. Thus we can say that Amazon.in provide highest varity of offers and paytm.com as lowest as comparison to other E-com sites. 

```python
#Combining numerical features
num_features=[features for features in cr.columns if cr[features].dtypes!="O"]
print(len(num_features))
cr[num_features].head()
```

```python
#Data Analysis on numerical Features:
for feature in num_features:  
    data=cr.copy()
    plt.figure(figsize=(10,5))
    data[feature].hist(bins=30)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

Analysis on Numerical Features:
1 Gender Analysis: Here we have analysed that rate of females is much higher than males for online shopping. from the graph we have observed that 
    females count is much higher than males count.
2 Age Analysis: From the graph we analysed that online shopping is higher for age between 31-40 years and least with age 51 and above years.
3 Pin Code Analysis: From the graph we analysed that Delhi Pincodes are maximum for online shopping and Bulandshar pincodes are count is least
4 Since how long you're shopping: Its observed from this dataset that people are shopping since 2-3 years and above 4 years are maximum and least with 1-2 years.
5 Online Purchase in last 1 year: Onlne purchase in last 1 year is maximum with 'less than 10 times' and minimum with 21-30 times.
6 While shopping mode of internet Access: Maximum with Mobile Internet and dial-up is least to access internet while shopping.
7 Screen size of mobile device: With help of graph visualization, we analysed that maximum number of users who are using others category for their mobile screen size that are around count of 135. After that 5.5 inch mobile screen size used by customers for online shopping that are around count of 100. Least count is for 5 inch and 4.7 inch of mobile device users.
